```python
# check if folder is mounted and if so, add it to the list of data sources to mount inside container
import os
path1 = "/path/to/data/source1"
path2 = "/path/to/data/source2"
path3 = "/path/to/data/source3"

if os.path.isdir(path1):
    # list content of folder and print first 5 entries
    print(f"Found data source: {path1}")
    print(f"Content of {path1}: {os.listdir(path1)[:5]}")
else:
    print(f"Data source not found: {path1}")

if os.path.isdir(path2):
    # list content of folder and print first 5 entries
    print(f"Found data source: {path2}")
    print(f"Content of {path2}: {os.listdir(path2)[:5]}")
else:
    print(f"Data source not found: {path2}")

if os.path.isdir(path3):
    # list content of folder and print first 5 entries
    print(f"Found data source: {path3}")
    print(f"Content of {path3}: {os.listdir(path3)[:5]}")
else:
    print(f"Data source not found: {path3}")
```
